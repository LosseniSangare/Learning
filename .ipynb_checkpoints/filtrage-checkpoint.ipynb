# Méthodes Utiles

```python
str(1234567890)[-4:]
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(13245689090)
```

```python
df['last_fort']= df['CC Number'].apply(last_four)
df
```

```python
def dolar(prix):
    if(prix<=10):
        return '$'
    elif (prix> 10 and prix <=30):
        return '$$'
    else :
        return '$$'
```

```python
df['dolar']= df['price_per_person'].apply(dolar)
df
```
